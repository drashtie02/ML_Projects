```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:\\Drashti\\Machine Learning\\Regression\\Salary_Data.csv")

df.head()
```

X = SELECT ALL THE ROWS,SELECT THE LAST COLUMNS EXCEPT LAST ONE

Y = SELECT ALL THE ROWS,SELCET FIRST COLUMN ONLY

```python
x = df.iloc[:,:-1]
y = df.iloc[:,:1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(x_train,y_train)
```

```python
x_pred = reg.predict(x_train)
y_pred = reg.predict(y_test)
```

```python
plt.scatter(x_train,y_train,color ="pink")
plt.plot(x_train,x_pred,color = "lavender")
plt.show()
```

```python
plt.scatter(x_test,y_test,color = "blue")
plt.plot(x_train,x_pred,color ="brown")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared
r2 = r2_score(y_test, y_pred)

r2
```
