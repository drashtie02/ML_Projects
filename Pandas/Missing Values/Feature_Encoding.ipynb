Apply function canbe done to series as well as DateFrame

Apply on DataFrame helps us dealing with multiple columns

function will receive all the rows

```python
df.lat.apply('sum')
```

```python
def fun(e):
    if e.long == '74.797371':
        return e.lat * 2
    else:
        return e.lat
```

```python
df.apply(fun,axis=1)
```

GROUPBY - it spits data into groups, a function is applied to each group seperately, combine results into a data structure

```python
df.groupby(['lat']).long.mean()
```

```python
df.groupby(['lat']).long.agg(['mean','min','max'])
```
